```python
def ngram_words(sentence: str, n:int):
    sentence = sentence.split()
    result = []
    sent_len = len(sentence)
    counter = 0
    while(counter+n<=sent_len):
        result.append(" ".join(sentence[counter:counter+n]))
        counter+=1
    return result
```

```python
def get_similarity(word,text_words):
  max_sim = -1
  for text_word in text_words:
    if text_word in model.wv.vocab and word in model.wv.vocab:
      sim = model.similarity(text_word,word)
      max_sim = max(max_sim,sim)
      if sim==1:
        break
    elif text_word not in model.wv.vocab and word in model.wv.vocab:
      sim = -1
    elif word not in model.wv.vocab:
      max_sim = -1
      break
  return max_sim
```

```python
def get_recomendation(text):
  text_words = text.split()
  digikala_words = digikala_samples_df.words.to_list()
  max_similarity = 0
  recom_item = 0
  recom_list = []
  for i in range(len(digikala_words)):
    similarity = 0
    for word in digikala_words[i]:
      similarity += get_similarity(word,text_words)
    max_similarity = max(max_similarity,similarity)
    recom_item = digikala_samples_df.category.iloc[i]
    recom_list.append((recom_item,similarity))
  result = sorted(recom_list,key=lambda x:x[1])[-5:]
  return result
```
