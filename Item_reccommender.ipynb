تابع زیر برای بدست آوردن ان-گرام‌ها است به این شکل که یک جمله و یک عدد به آن‌ میدهیم و این تابع کلمات به آن تعداد که کنار هم هستند را برمیگرداند مقلا برای بایگرام باید عدد دو بدهیم.

```python
def ngram_words(sentence: str, n:int):
    sentence = sentence.split()
    result = []
    sent_len = len(sentence)
    counter = 0
    while(counter+n<=sent_len):
        result.append(" ".join(sentence[counter:counter+n]))
        counter+=1
    return result
```

# ساخت سیستم پیشنهاد دهنده‌ی اول بر مبنای 
# jaccard similarity   

الان با کمک دو تابع زیر میخواهیم میزان شباهت بین دو متن را به شکل خیلیی ساده بدست بیاوریم و سپس بر اساس میزان شباهت به هر متن ۵ کالای اول را پیشنهاد دهیم.
تابع اول بین دو دیشکنری از فرکانس‌ها تعداد کلمات مشترک و میزان تکرار آن‌ها را بدست آورده و تقسیم بر تعداد کل کلمات یکتای بین دو متن میکند. این کار تقریبا شبیه به معیار 
jaccard similarity
می‌باشد.

```python
def convert_text(text):
  text_dict = {}
  text_dict["words"] = list(map(lemmatizer.lemmatize,[i for i in text.split() if i not in stopwords]))
  text_dict["freq_dict"] = dict(Counter(text_dict["words"]))
  text_dict["bigram_words"] = ngram_words(" ".join(text_dict["words"]),2)
  text_dict["bigram_freq_dict"] = dict(Counter(text_dict["bigram_words"]))
  return text_dict
```

در اینجا برای بیان شباهت بین دو کلمه ۲ راه موجود بود یکی اینکه خیلی ساده چک کنیم چه تعداد کلمه‌ی مشترک دارند و دیگری اینکه میزان شباهت کلمات آن‌ها را بسنجیم. مورد اول به مراتب ساده‌تر و سریع‌تر بود ولی میزان شباهت کمتری را نشان میداد اما روش دوم زمان‌گیر و کند بود زیرا باید تک تک کلمات را ۲ به ۲ بررسی میکردیم در عوض دقت بیشتری داشت و میزان شباهت بالاتری به ما میداد. نحوه عملکرد هم به این شکل بود که اگر میزان شباهت بیشتر از ۰.۷۵ بود تعداد تکرار آن کلمه را ضرب در میزان مشابهت کرده و با میزان شباهت کلی متن جمع میکردیم.

```python
def calculate_dicts_similarity(text_dict:Dict, item_dict:Dict):
  similarity = 0
  for tk in text_dict.keys():
    if tk in item_dict.keys():
      sim = 1
    # for ik in item_dict.keys():
      # sim = difflib.SequenceMatcher(None,ik,tk).ratio()
      # if sim > 0.75:
      similarity += (text_dict[tk]*sim)
  similarity/=(len(text_dict)+len(item_dict))
  return similarity
```

بدست می‌آید که در واقع کسینوس زاویه‌ی بین ۲بردار را حساب میکند و هرچه ۲بردار به هم نزدیک‌تر باشن(۲کلمه شبیه‌تر باشند) به یک نزدیک‌تر است و اگر مخالف هم باشند به -۱ نزدیک‌تر است.
حال تابعی مانند قسمت قبل پیاده میکنیم ولی اینبار ۲ به ۲ کلمات را چک میکنیم و ماکسیمم شباهت کسینوسی را برای هر کلمه بدست می‌آوریم و با هم جمع میکنیم در نهایت پنج کالا با بیشترین شباهت به متن را خروجی میدهیم.

```python
import gensim
from gensim.models.word2vec import Word2Vec
```

```python
wiki_texts = [text for text in wiki_df.Text]
digikala_texts = [text for text in digikala_samples_df.words]
```

```python
texts = [word_tokenize(d_text) for d_text in map(" ".join,digikala_texts)]
texts.extend([word_tokenize(w_text) for w_text in wiki_texts])
```

```python
model = Word2Vec(texts, size=64, window=10, min_count=5, seed=42, workers=5)
```

```python
len(model.wv.vocab.keys())
```

```python
def get_similarity(word,text_words):
  max_sim = -1
  for text_word in text_words:
    if text_word in model.wv.vocab and word in model.wv.vocab:
      sim = model.similarity(text_word,word)
      max_sim = max(max_sim,sim)
      if sim==1:
        break
    elif text_word not in model.wv.vocab and word in model.wv.vocab:
      sim = -1
    elif word not in model.wv.vocab:
      max_sim = -1
      break
  return max_sim
```

```python
def get_recomendation(text):
  text_words = text.split()
  digikala_words = digikala_samples_df.words.to_list()
  max_similarity = 0
  recom_item = 0
  recom_list = []
  for i in range(len(digikala_words)):
    similarity = 0
    for word in digikala_words[i]:
      similarity += get_similarity(word,text_words)
    max_similarity = max(max_similarity,similarity)
    recom_item = digikala_samples_df.category.iloc[i]
    recom_list.append((recom_item,similarity))
  result = sorted(recom_list,key=lambda x:x[1])[-5:]
  return result
```

در این قسمت میخواهیم به کمک شباهت کسینوسی شباهت میان همه‌ی ۲۱هزار متن موجود در دیتافریم را محاسبه کنیم و در یک ماتریس جدا ذخیره کنیم. این عمل کمی طول میکشد و به رم و حافظه‌ی نسبتا زیادی احتیاج دارد ولی باعث می‌شود که پس از محاسبه و ساخت این ماتریس دیگر عمل پیشنهاد دادن در کمتر از ۱ ثانیه انجام بگیرد.

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

حال که ماتریس را ساختیم تابع زیر را مینویسیم.

تابع زیر یک عدد به عنوان ورودی میگیرد این عدد باید شماره‌ی متن ویکی‌پدیا در دیتافریم ویکی‌پدیا باشد(۲۰هزار متن در این دیتافریم موجود است)
با دادن شماره‌ی این متن تابع زیر ۱۰تا از شبیه‌ترین کالاها به متن را پیدا کرده و پیشنهاد میدهد.

در بخشی از کد از فیلتر استفاده کرده‌ایم و مشخص شده که موارد انتخابی باید ایندکس بیشتر از ۲۰هزار داشته‌باشند. دلیل این امر آن است که ۲۰هزار مورد اول درواقع متن‌های ویکی‌پدیا هستند و ما نمیخواهیم مورد پیشنهادی هم متنی از ویکی‌پدیا باشد و باید یک کالا باشد که کالاها هم در سطر ۲۰هزار به بعد قرار دارند.

اما اگر بخواهیم متنی از ویکی‌پدیا را به عنوان متن شبیه به متن فعلی پیشنهاد دهیم میتوان ورودی دوم را ۰ گذاشت تا بین همه‌ی موارد پیشنهاد دهد.

```python
def get_recomendation(wiki_text_num:int, start=20000):
  sim_scores = list(enumerate(cosine_sim[wiki_text_num]))
  sim_scores = filter(lambda x:x[0]>=start,sim_scores)
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[0:10]
  # recom_items = all_texts_df.loc[all_texts_df.index.isin(list(map(lambda x:x[0],sim_scores)))]
  # recom_items["similarity"] = 
  sim_scores_list = []
  for sim_score in sim_scores:
    sim_scores_list.append((all_texts_df["texts"].iloc[sim_score[0]],sim_score[1]))
  return sim_scores_list
```
